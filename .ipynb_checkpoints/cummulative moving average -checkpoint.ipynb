```python
import yfinance as yf
import pandas as pd
from datetime import datetime
```

```python
df_tesla = yf.download("TSLA", period="max")

print(df_tesla.head())
```

```python
#simple moving average
df_tesla["Open"].plot(figsize=(15,6))
```

```python
df_tesla['Open:30 days rolling']=df_tesla[('Open', 'TSLA')].rolling(window=30,min_periods=1).mean()
```

```python
df_tesla[['Open','Open:30 days rolling']].plot(xlim=['2019-06-29',"2020-06-29"],figsize=(12,5))
```

```python
df_tesla['Open:10 days rolling']=df_tesla[('Open', 'TSLA')].rolling(window=10,min_periods=1).mean()
```

```python
df_tesla['Open:20 days rolling']=df_tesla[('Open', 'TSLA')].rolling(window=20,min_periods=1).mean()
```

```python
df_tesla[['Open','Open:30 days rolling','Open:10 days rolling','Open:20 days rolling']].plot(xlim=['2019-06-29',"2020-06-29"],ylim=[0,100],figsize=(12,5))
```

```python
#expanding 
#cummulative moving average 
df_tesla["Open"].expanding().mean ().plot(figsize=(10,5))
```
